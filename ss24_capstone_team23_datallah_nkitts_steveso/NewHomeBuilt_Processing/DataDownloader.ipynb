(1995-2019) Historical Permits by Metro data following this url pattern:
- https://www.census.gov/construction/bps/txt/tb3u201001.txt (2010.01)
- https://www.census.gov/construction/bps/txt/tb3u201002.txt (2010.02)

```python
# import libraries
import requests
import pandas as pd
from google.colab import drive
```

```python
# mount G drive
drive.mount('/content/drive')
```

```python
def download_and_save_historical_data(year, month):
  url = f'https://www.census.gov/construction/bps/txt/tb3u{year}{month:02d}.txt' # Assuming url pattern
  filename = f'NewHomePermit-{year}-{month:02d}.txt'
  response = requests.get(url)

  if response.status_code == 200:
    with open(f'/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/NewHomePermit/{filename}', 'wb') as f:
      f.write(response.content)
    print(f'File {filename} downloaded and saved to Google Drive.')
  else:
    print(f'Failed to download {filename}. Status code: {response.status_code}')

# Example usage:
download_and_save_historical_data(2010, 1)
```

```python
for year in range(2010, 2020):
  for month in range(1, 13):
    download_and_save_historical_data(year, month)
```

(2019-present) Current Permits by Metro data following this url pattern:
- https://www.census.gov/construction/bps/xls/msamonthly_201911.xls (2019.11)
- https://www.census.gov/construction/bps/xls/msamonthly_201912.xls (2019.12)

```python
def download_and_save_current_data(year, month):
  url = f'https://www.census.gov/construction/bps/xls/msamonthly_{year}{month:02d}.xls' # Assuming url pattern
  filename = f'NewHomePermit-{year}-{month:02d}.xls'
  response = requests.get(url)

  if response.status_code == 200:
    with open(f'/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/NewHomePermit/{filename}', 'wb') as f:
      f.write(response.content)
    print(f'File {filename} downloaded and saved to Google Drive.')
  else:
    print(f'Failed to download {filename}. Status code: {response.status_code}')

# Example usage:
download_and_save_historical_data(2019, 11)
```

```python
for year in range(2019, 2024):
  if year == 2019:
    for month in range(11, 13):
      download_and_save_current_data(year, month)
  else:
    for month in range(1, 13):
      download_and_save_current_data(year, month)
```
