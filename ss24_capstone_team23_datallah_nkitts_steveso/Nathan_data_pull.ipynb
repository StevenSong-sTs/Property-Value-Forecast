```python
Census_api_key = False 

if Census_api_key is False:
    Census_api_key = os.environ.get("Census_api_key")
else : 
    Census_api_key = False

#base url
#resconst
#vip
#census_api_url="https://api.census.gov/data/timeseries/bps?get=cell_value,time_slot_id&"
census_api_url="https://api.census.gov/data/timeseries/bps?get=cell_value,time_slot_id&"


#placeholders for required parameters
error_data = ""
seasonally_adj = ""
category_code = ""
data_type_code = ""
geo_level_code = ""
for_param = ""
in_param =""
#time = "2022"

#param dictionary
params = dict()
params["error_data"] = error_data
params["seasonally_adj"] = seasonally_adj
params["category_code"] = category_code
params["data_type_code"] = data_type_code
params["geo_level_code"] = geo_level_code
params["for"] = for_param
#params["in"] = in_param
params["time"] = time

#if key exists add key to end of dictionary
if Census_api_key is not False: params["key"] = Census_api_key

#create query string
url = census_api_url + urllib.parse.urlencode(params,doseq=True)
print('Retrieving', url)

#Pull Data
Cenus_api_data = urllib.request.urlopen(url)
Census_data = Cenus_api_data.read().decode()
print('Retrieved', len(Census_data), 'charachters')

#transfer data into JSON
try:
    js = json.loads(Census_data)
except:
    js=None

if not js:
    print('==== Failure To Retrieve ====')
    print(Census_data)


print(json.dumps(js, indent=4))
```

```python
import requests
import json
import prettytable

prefix='WM'
season = 'S' # S for Adjusted, U for unadjusted
area = '0042660' #Seattle-Tacoma-Bellvue
owner= '1' #1:Federal, 2:State, 3:Local, 4:Private
estimate = '02' #01:Employment 02:Number of establishments, 03:Average weekly wage, 04:Total wages, 05:Taxable wages, 06:Contributions, 07:Contribution rates, 08:Total remuneration, 09:Administrative funding
industry = '000000'
occupation = '110000'
job_char = '00'
level='10'

ID=prefix+season+area+owner+estimate+industry+occupation+job_char+level
print(ID)

headers = {'Content-type': 'application/json'}
data = json.dumps({
    "seriesid": [f'{ID}'],
    "startyear":"2010",
    "endyear":"2024",
    "registrationkey": BLS_api_key
      })
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
for series in json_data['Results']['series']:
    print(series)
    x=prettytable.PrettyTable(["series id","year","period","value","footnotes"])
    seriesId = series['seriesID']
    for item in series['data']:
        year = item['year']
        period = item['period']
        value = item['value']
        footnotes=""
        for footnote in item['footnotes']:
            if footnote:
                footnotes = footnotes + footnote['text'] + ','
        if 'M01' <= period <= 'M12':
            x.add_row([seriesId,year,period,value,footnotes[0:-1]])
    output = open(seriesId + '.txt','w')
    output.write (x.get_string())
    output.close()
    print(json_data)
```
