## 1. Aggregate to monthly count

```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
crime_df = pd.read_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/CrimeData/Chicago/Chicago-CrimeData-2001-Presen.csv")
```

```python
crime_df.shape
```

```python
crime_df["Primary Type"].unique()
```

```python
## Crime types we care:
violent_crimes = [
    'BATTERY', 'ASSAULT', 'SEX OFFENSE', 'CRIM SEXUAL ASSAULT', 'ROBBERY',
    'CRIMINAL SEXUAL ASSAULT', 'STALKING', 'HOMICIDE', 'KIDNAPPING',
    'INTIMIDATION', 'DOMESTIC VIOLENCE', 'WEAPONS VIOLATION'
]

property_crimes = [
    'THEFT', 'BURGLARY', 'MOTOR VEHICLE THEFT', 'CRIMINAL DAMAGE',
    'ARSON', 'CRIMINAL TRESPASS'
]
```

```python

violent_crime_df = crime_df[crime_df["Primary Type"].isin(violent_crimes)]
property_crime_df = crime_df[crime_df["Primary Type"].isin(property_crimes)]
```

```python
print(f"shape of violent crime df: {violent_crime_df.shape}")
print(f"shape of property crime df: {property_crime_df.shape}")
```

```python

violent_crime_df["Date"] = pd.to_datetime(violent_crime_df["Date"])
property_crime_df["Date"] = pd.to_datetime(property_crime_df["Date"])

```

```python

violent_crimes_by_month = violent_crime_df.groupby(pd.Grouper(key='Date', freq='M')).size().reset_index(name='NumberOfIncident')
violent_crimes_by_month['Date'] = violent_crimes_by_month['Date'].dt.strftime('%Y-%m')

property_crimes_by_month = property_crime_df.groupby(pd.Grouper(key='Date', freq='M')).size().reset_index(name='NumberOfIncident')
property_crimes_by_month['Date'] = property_crimes_by_month['Date'].dt.strftime('%Y-%m')
```

```python
violent_crimes_by_month.to_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/chicago_violent_crimes_by_month.csv", index=False)
property_crimes_by_month.to_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/chicago_property_crimes_by_month.csv", index=False)
```

## Get Population Data and Calcualte Crime Rate / 100,000 people

```python
import pandas as pd

population_df = pd.read_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/Population/Chicago_Population_2010-2022.csv")
violent_crimes_by_month_df = pd.read_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/chicago_violent_crimes_by_month.csv")
property_crimes_by_month_df = pd.read_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/chicago_property_crimes_by_month.csv")
```

```python
violent_crimes_by_month_df.head(12)
```

```python
violent_crimes_by_year = violent_crimes_by_month_df.groupby(violent_crimes_by_month_df['Date'].str[:4]).agg(
    {'NumberOfIncident': 'sum'}).reset_index()
violent_crimes_by_year.columns = ['Year', 'NumberOfIncident']

property_crimes_by_year = property_crimes_by_month_df.groupby(property_crimes_by_month_df['Date'].str[:4]).agg(
    {'NumberOfIncident': 'sum'}).reset_index()
property_crimes_by_year.columns = ['Year', 'NumberOfIncident']
```
