# Seattle Crime Data Processing

```python
import pandas as pd

crime_df = pd.read_csv('/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/CrimeData/Seattle/Seattle_CrimeData_2008_2024.csv')
```

```python
crime_df["Crime Against Category"].unique()
```

```python
crime_df = crime_df[crime_df["Crime Against Category"] != "NOT_A_CRIME"]
```

```python
crime_df["Offense Parent Group"].unique()
```

```python
violent_crimes = [
    "KIDNAPPING/ABDUCTION",
    "SEX OFFENSES",
    "HUMAN TRAFFICKING",
    "HOMICIDE OFFENSES",
    "ASSAULT OFFENSES",
    "WEAPON LAW VIOLATIONS",
    "ANIMAL CRUELTY"
]

violent_crime_df = crime_df[crime_df["Offense Parent Group"].isin(violent_crimes)]
```

```python
property_crime_df = crime_df[crime_df["Crime Against Category"] == "PROPERTY"]
```

```python
violent_crime_df["Crime Against Category"].unique()
```

```python
violent_crime_df["Report DateTime"] = pd.to_datetime(violent_crime_df["Report DateTime"])
property_crime_df["Report DateTime"] = pd.to_datetime(property_crime_df["Report DateTime"])
```

```python
violent_crimes_by_month = violent_crime_df.groupby(pd.Grouper(key='Report DateTime', freq='M')).size().reset_index(name='NumberOfIncident')
violent_crimes_by_month['Report DateTime'] = violent_crimes_by_month['Report DateTime'].dt.strftime('%Y-%m')

property_crimes_by_month = property_crime_df.groupby(pd.Grouper(key='Report DateTime', freq='M')).size().reset_index(name='NumberOfIncident')
property_crimes_by_month['Report DateTime'] = property_crimes_by_month['Report DateTime'].dt.strftime('%Y-%m')
```

```python
violent_crimes_by_month
```

```python
# Rename column
violent_crimes_by_month = violent_crimes_by_month.rename(columns={'Report DateTime': 'Date'})
property_crimes_by_month = property_crimes_by_month.rename(columns={'Report DateTime': 'Date'})
```

```python
violent_crimes_by_month.to_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/Crime/seattle_violent_crimes_by_month.csv", index=False)
property_crimes_by_month.to_csv("/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/Crime/seattle_property_crimes_by_month.csv", index=False)
```
