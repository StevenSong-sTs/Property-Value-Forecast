```python
hud_path = '/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/hud_40th_percentiles/FMR_All_1983_2024_revised (1).csv'
```

```python
import pandas as pd
pd.options.display.max_columns = None

import numpy as np
import impyute as impy


metros={
    'New York-Newark-Jersey City, NY-NJ-PA Metro Area':{'Metro_Prefix':'METRO35620'},
    'Los Angeles-Long Beach-Anaheim, CA Metro Area':{'Metro_Prefix':'METRO31080'},
    'Chicago-Naperville-Elgin, IL-IN-WI Metro Area':{'Metro_Prefix':'METRO16980'},
    'Dallas-Fort Worth-Arlington, TX Metro Area':{'Metro_Prefix':'METRO19100'},
    'Houston-The Woodlands-Sugar Land, TX Metro Area':{'Metro_Prefix':'METRO26420'},
    'Washington-Arlington-Alexandria, DC-VA-MD-WV Metro Area':{'Metro_Prefix':'METRO47900'},
    'Philadelphia-Camden-Wilmington, PA-NJ-DE-MD Metro Area':{'Metro_Prefix':'METRO37980'},
    'Atlanta-Sandy Springs-Alpharetta, GA Metro Area':{'Metro_Prefix':'METRO12060'},
    'Miami-Fort Lauderdale-Pompano Beach, FL Metro':{'Metro_Prefix':'METRO33100'},
    'Phoenix-Mesa-Chandler, AZ Metro Area':{'Metro_Prefix':'METRO38060'},
    'Boston-Cambridge-Newton, MA-NH Metro Area':{'Metro_Prefix':'METRO14460'},
    'Riverside-San Bernardino-Ontario, CA Metro Area':{'Metro_Prefix':'METRO40140'},
    'San Francisco-Oakland-Berkeley, CA Metro Area':{'Metro_Prefix':'METRO41860'},
    'Detroit-Warren-Dearborn, MI Metro Area':{'Metro_Prefix':'METRO19820'},
    'Seattle-Tacoma-Bellevue, WA Metro Area':{'Metro_Prefix':'METRO42660'},
    }

metro_prefixes =[]
for key in metros.keys():
    metro_prefixes.append(metros[key]['Metro_Prefix'])
```

```python
hud=pd.read_csv(hud_path, encoding = "ISO-8859-1")

hud = hud[hud['cbsasub'].str.startswith(tuple(metro_prefixes))]
hud['fmr_overaching'] = hud['cbsasub'].str[0:10]
hud['fmr24'] = hud['fmr23']

display(hud.head())
```

```python
import pandas as pd
import numpy as np


filter_col = ['fmr_overaching'] + sorted([col for col in hud.columns if col.startswith('msa')])
msa_df = hud[filter_col].copy()


for col in msa_df.columns:
    if col == 'fmr_overaching':
        continue
    if msa_df[col].dtype == object:
        msa_df.loc[:, col] = msa_df[col].str.extract(r'.*?([0-9]+)$')[0]

    msa_df.loc[:, col] = pd.to_numeric(msa_df[col], errors='coerce')
    msa_df.loc[msa_df[col].isna(), col] = 0
    msa_df.loc[:, col] = msa_df[col].astype(int)

# Transpose the DataFrame, setting 'fmr_overaching' as the index
msa_df = msa_df.set_index('fmr_overaching').T
rows_with_zero = msa_df.isin([0]).any(axis=1)
msa_df = msa_df[~rows_with_zero]

# Melt the DataFrame, keeping 'fmr_overaching' as identifier and getting 'area_code' from values
msa_df = pd.melt(msa_df.T.reset_index(), id_vars=['fmr_overaching'], var_name='Variable', value_name='area_code')
#msa_df = msa_df[~msa_df['area_code'].isin([9999, 10000])]
msa_df = msa_df.drop('Variable', axis=1).drop_duplicates()
msa_df = msa_df.rename({'fmr_overaching': 'metro_code'}, axis=1)


# Create a reverse mapping dictionary from metro codes to metro names
reverse_mapping = {value['Metro_Prefix']: key for key, value in metros.items()}
msa_df['metro_name'] = msa_df['metro_code'].map(reverse_mapping)


msa_df.to_json('/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/metro_areas.json', orient='records', lines = True)
```

```python
year_codes = [f"{i:02}" for i in range(25)]

fmr_records = []

for metro, metro_data in metros.items():
    metro_prefix = metro_data['Metro_Prefix']
    metro_fmr_areas = list(hud[hud['fmr_overaching'] == metro_prefix]['fmr_area'].unique())

    for area in metro_fmr_areas:
        hud_area_filtered = hud[hud['fmr_area'] == area]
        hud_area_name = hud_area_filtered["areaname"].iloc[0]

        for code in year_codes:
            year = f'20{code}'
            fmr_data = hud_area_filtered.iloc[0]


            fmr_record = {
                "metro_name": str(metro),
                "metro_code": str(metro_prefix),
                "area_name": str(hud_area_name),
                "area_code": str(int(area)),
                "year": int(year),
                "percentile": int(fmr_data[f'fmr{code}']),
                "studio": int(fmr_data[f'fmr{code}_0']),
                "1_bedroom": int(fmr_data[f'fmr{code}_1']),
                "2_bedroom": int(fmr_data[f'fmr{code}_2']),
                "3_bedroom": int(fmr_data[f'fmr{code}_3']),
                "4_bedroom": int(fmr_data[f'fmr{code}_4']),

            }

            if year in ['2000', '2010', '2017']:
                fmr_record['population'] = float(hud_area_filtered[f'pop{year}'].sum())
            else:
                fmr_record['population'] =np.NaN
            fmr_records.append(fmr_record)





# Convert records to a DataFrame directly
fmr_df = pd.DataFrame(fmr_records)
fmr_df
```

```python
for area in fmr_df['area_code']:
    area_mask = fmr_df['area_code'] == area
    area_data = fmr_df[area_mask]
    pop = area_data[['population']]
    fmr_df.loc[area_mask, 'pop_est'] = impy.em(pop.values, loops = 50)

fmr_df['helper'] =1
data_col_mask= ['studio','1_bedroom','2_bedroom','3_bedroom','4_bedroom']

for col in data_col_mask:
    fmr_df[col] = fmr_df[col]* fmr_df['pop_est']

fmr_df=fmr_df.groupby(by=['metro_name', 'metro_code', 'year']).sum().reset_index()

for col in data_col_mask:
    fmr_df[col] = fmr_df[col]/fmr_df['pop_est']

fmr_df['percentile'] = fmr_df['percentile']/fmr_df['helper']

fmr_df
```

```python
fmr_final_cols=['metro_name', 'metro_code', 'year','percentile','studio','1_bedroom','2_bedroom','3_bedroom','4_bedroom']
fmr_df[fmr_final_cols].to_json('/content/drive/MyDrive/MADS Capstone Team 23/Data/processed/hud_40th_percentiles/hud_data.json', orient='records', lines = True)
```
