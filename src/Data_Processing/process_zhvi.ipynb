## Total Population in All Metropolitans and Micropolitans 2021
Find out the top N Metropolitans with largest population to determine the interested regions.


```python
NUMBER_OF_CITY = 15
```

```python
import pandas as pd

population_df = pd.read_csv('/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/total_population_2022.csv')

filtered_population_df = population_df[['NAME', 'B01003_001E']]

cleaned_df = filtered_population_df.iloc[1:]

cleaned_df = cleaned_df.astype({
    'NAME': 'str',
    'B01003_001E': 'int64'
})

cleaned_df.rename(columns={'NAME': 'Name', 'B01003_001E': 'Population'}, inplace=True)

cleaned_df.sort_values(by='Population', ascending=False, inplace=True)

cleaned_df = cleaned_df.drop_duplicates()

top15_metro_df = cleaned_df.head(NUMBER_OF_CITY)

top15_metro_df
```

```python
top15_metro_df_copy = top15_metro_df.copy()

top15_metro_df_copy['RegionName'] = ''

for index, row in top15_metro_df_copy.iterrows():
  name_string = row['Name']

  split_1 = name_string.split(',')

  first_part = split_1[0]

  split_2 = first_part.split('-')

  metro_name = split_2[0]

  second_part = split_1[1]

  state_abbrev = second_part[:3]

  region_name = metro_name + ',' + state_abbrev

  top15_metro_df_copy.loc[index, 'RegionName'] = region_name

top15_metro_df = top15_metro_df_copy

top15_metro_df
```

## Zillow Home Value Index for the 15 Metros we will be focusing on

```python
zhvi_df = pd.read_csv('/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/zhvi_metro.csv')

zhvi_df.head()

merged_df = pd.merge(zhvi_df, top15_metro_df, on='RegionName')

processed_zhvi_df = merged_df.drop(columns=['Name', 'Population'])

processed_zhvi_df = processed_zhvi_df.melt(id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName'])

processed_zhvi_df.rename(columns={'variable': 'TimePeriod', 'value': 'ZHVI'}, inplace=True)

processed_zhvi_df['TimePeriod'] = pd.to_datetime(processed_zhvi_df['TimePeriod'])

processed_zhvi_df
```

```python
processed_zhvi_df.to_csv('/content/drive/MyDrive/MADS Capstone Team 23/Data//processed/processed_zhvi_df.csv')
```

## Prepare the Data for Merging

```python
import pandas as pd

cities = ['New York, NY', 'Los Angeles, CA', 'Chicago, IL', 'Dallas, TX', 'Seattle, WA']

zhvi_df = pd.read_csv('/content/drive/MyDrive/MADS Capstone Team 23/Data/raw/zhvi_metro.csv')

filtered_zhvi_df = zhvi_df[zhvi_df['RegionName'].isin(cities)]

filtered_zhvi_df = filtered_zhvi_df.drop(columns=['RegionID', 'SizeRank', "RegionType", "StateName"])

filtered_zhvi_df["City"] = filtered_zhvi_df["RegionName"].str.split(",").str[0]

filtered_zhvi_df = filtered_zhvi_df.drop(columns=['RegionName'])
```

```python
processed_zhvi_df = filtered_zhvi_df.melt(id_vars=['City'])
processed_zhvi_df.rename(columns={'variable': 'Date', 'value': 'ZHVI'}, inplace=True)
processed_zhvi_df['Date'] = pd.to_datetime(processed_zhvi_df['Date'])
processed_zhvi_df = processed_zhvi_df[(processed_zhvi_df['Date'] >= '2010-01') & (processed_zhvi_df['Date'] <= '2023-01')]
processed_zhvi_df['Date'] = processed_zhvi_df['Date'].dt.strftime('%Y-%m')
```
